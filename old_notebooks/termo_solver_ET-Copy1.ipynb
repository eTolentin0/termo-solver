```python
import unicodedata
import re

def remove_acentos(string):
    normalized = unicodedata.normalize('NFD', string)
    return re.sub(r'[\u0300-\u036f]', '', normalized).casefold()
```

```python
def conta_vogais(palavra):
    
    contador = 0
    vogais = 'aeiouAEIOU'
    allowed = frozenset('aeiouAEIOU')
    
    for letra in palavra:
        if letra in vogais:
            contador += 1
    vogais_unicas = len(allowed.intersection(palavra))
    
    return vogais_unicas, contador


conta_vogais('perde')    

```

```python
def contem_letras(letras, palavra):
    contador = 0

    for letra in letras: #estava errando aqui ao referencias palavras e não letras que quero que cheque
        if letra in palavra:
            contador += 1

            
    return contador == len(letras)


def tira_palavras_com_letras(letras, palavra):
    contador = 0
    
    for letra in letras:
        if letra in palavra:
            contador += 1
            
    return contador > 0

def lista_de_palavras_possiveis(letras_existentes, letras_n_existentes, lista_de_palavras):
    listinha = []
    

    for palavra in lista_de_palavras:
        if contem_letras(letras_existentes, palavra) == True and tira_palavras_com_letras(letras_n_existentes, palavra) == False:
            listinha.append(palavra)
    
    return listinha
```
